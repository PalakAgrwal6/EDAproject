```python
#Baidu : Quarterly Revenue, Expenses and Income
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/Baidu_ir - Sheet1.csv")
df
```

```python

data = {
    'Quarter Report Date': ['2023-01-01', '2023-04-01', '2023-07-01', '2023-10-01'],
    'Baidu Core total cost and expenses': [500, 700, 650, 800],
    'iQIYI total cost and expenses': [300, 400, 350, 450]
}
df = pd.DataFrame(data)
df['Quarter Report Date'] = pd.to_datetime(df['Quarter Report Date'])
df.set_index('Quarter Report Date', inplace=True)

df[['Baidu Core total cost and expenses', 'iQIYI total cost and expenses']].plot(figsize=(12, 6))
plt.title('Revenue and Expenses Over Time')
plt.xlabel('Quarter Report Date')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
print(df.columns)
```

```python
import seaborn as sns

# Select financial columns
financial_data = df[['Baidu Core Revenu', 'Baidu Core total cost and expenses', 'Opearting Income Baidu Core']]

# Heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(financial_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation of Financial Metrics')
plt.show()
```

```python
# Histogram for revenue distribution
df['Baidu Core Revenu'].plot(kind='hist', bins=15, color='orange', edgecolor='black', figsize=(8, 6))
plt.title('Revenue Distribution')
plt.xlabel('Baidu Core Revenu')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot for Revenue vs Expense
plt.figure(figsize=(8, 6))
plt.scatter(df['Baidu Core Revenu'], df['Baidu Core total cost and expenses'], alpha=0.7, color='purple')
plt.title('Revenue vs Expense')
plt.xlabel('Baidu Core Revenu')
plt.ylabel('Baidu Core total cost and expenses')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Revenue vs Expense
plt.figure(figsize=(8, 6))
plt.scatter(df['Baidu Core Revenu'], df['iQIYI Revenue'], alpha=0.7, color='purple')
plt.title(' Baidu Revenue vs iQIYI Revenue')
plt.xlabel('Baidu Core Revenu')
plt.ylabel('iQIYI Revenue')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Revenue vs Expense
plt.figure(figsize=(8, 6))
plt.scatter(df['Quarter Report Date'], df['Opearting Income Baidu Core'], alpha=0.7, color='purple')
plt.title('Revenue vs Expense')
plt.xlabel('Quarter Report Date')
plt.ylabel('Opearting Income Baidu Core')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Revenue vs Expense
plt.figure(figsize=(8, 6))
plt.scatter(df['Opearting Income Baidu Core'], df['iQIYI Opearting Income'], alpha=0.7, color='purple')
plt.title(' Opearting Income Baidu Core vs iQIYI Opearting Income')
plt.xlabel('Opearting Income Baidu Core')
plt.ylabel('iQIYI Opearting Income')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Revenue vs Expense
plt.figure(figsize=(8, 6))
plt.scatter(df['Quarter Report Date'], df['iQIYI Opearting Income'], alpha=0.7, color='purple')
plt.title('Revenue vs Expense')
plt.xlabel('Quarter Report Date')
plt.ylabel('iQIYI Opearting Income')
plt.grid(True)
plt.show()
```

```python
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):

        plt.figure(figsize=(8, 8))
        df[column].value_counts().plot(
            kind='pie',
            autopct='%1.1f%%',
            startangle=140,
            title=f"Pie Chart for {column}"
        )
        plt.ylabel('')
        plt.show()
    else:

        if len(df[column].unique()) < 10:
            for numeric_col in df.select_dtypes(include='number').columns:
                plt.figure(figsize=(8, 8))
                df.groupby(column)[numeric_col].sum().plot(
                    kind='pie',
                    autopct='%1.1f%%',
                    startangle=140,
                    title=f"Pie Chart for {numeric_col} by {column}"
                )
                plt.ylabel('')
                plt.show()
```
